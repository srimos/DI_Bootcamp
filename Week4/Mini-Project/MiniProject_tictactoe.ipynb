```python
def display_board(tictactoeboard):
    for row in tictactoeboard:
        print(row)

def player_input(tictactoeboard,player):
    while True:
        rplayer=int(input("Enter row:"))
        cplayer=int(input("Enter column:"))
        select=tictactoeboard[rplayer][cplayer]
        if select==' ':
            if player=='x':
                tictactoeboard[rplayer][cplayer]='X'
                break
            else:   
                tictactoeboard[rplayer][cplayer]='O'
                break
        else:
            print("This cell has already been chosen")

def check_winX(tictactoeboard):
    if tictactoeboard[1][1]=='X' and tictactoeboard[1][2]=='X' and tictactoeboard[1][3]=='X':
        return True
    elif tictactoeboard[2][1]=='X' and tictactoeboard[2][2]=='X' and tictactoeboard[2][3]=='X':
        return True
    elif tictactoeboard[3][1]=='X' and tictactoeboard[3][2]=='X' and tictactoeboard[3][3]=='X':
        return True
    elif tictactoeboard[1][1]=='X' and tictactoeboard[2][1]=='X' and tictactoeboard[3][1]=='X':
        return True
    elif tictactoeboard[1][2]=='X' and tictactoeboard[2][2]=='X' and tictactoeboard[3][2]=='X':
        return True
    elif tictactoeboard[1][3]=='X' and tictactoeboard[2][3]=='X' and tictactoeboard[3][3]=='X':
        return True
    elif tictactoeboard[1][1]=='X' and tictactoeboard[2][2]=='X' and tictactoeboard[3][3]=='X':
        return True
    elif tictactoeboard[3][1]=='X' and tictactoeboard[2][2]=='X' and tictactoeboard[1][3]=='X':
        return True
    else:
        return False

def check_winO(tictactoeboard):
    if tictactoeboard[1][1]=='O' and tictactoeboard[1][2]=='O' and tictactoeboard[1][3]=='O':
        return True
    elif tictactoeboard[2][1]=='O' and tictactoeboard[2][2]=='O' and tictactoeboard[2][3]=='O':
        return True
    elif tictactoeboard[3][1]=='O' and tictactoeboard[3][2]=='O' and tictactoeboard[3][3]=='O':
        return True
    elif tictactoeboard[1][1]=='O' and tictactoeboard[2][1]=='O' and tictactoeboard[3][1]=='O':
        return True
    elif tictactoeboard[1][2]=='O' and tictactoeboard[2][2]=='O' and tictactoeboard[3][2]=='O':
        return True
    elif tictactoeboard[1][3]=='O' and tictactoeboard[2][3]=='O' and tictactoeboard[3][3]=='O':
        return True
    elif tictactoeboard[1][1]=='O' and tictactoeboard[2][2]=='O' and tictactoeboard[3][3]=='O':
        return True
    elif tictactoeboard[3][1]=='O' and tictactoeboard[2][2]=='O' and tictactoeboard[1][3]=='O':
        return True
    else:
        return False

def check_tie(tictactoeboard,turn):
    if turn==9 and tictactoeboard[1][1]!=' ' and tictactoeboard[1][2]!=' ' and tictactoeboard[1][3]!=' ' and tictactoeboard[2][1]!=' ' and tictactoeboard[2][2]!=' ' and tictactoeboard[2][3]!=' ' and tictactoeboard[3][1]!=' ' and tictactoeboard[3][2]!=' ' and tictactoeboard[3][3]!=' ':
        return True
    else:
        return False

def play():
    print("Welcome to TIC TAC TOE!")
    player=input("Choose your player (x/o)")
    tictactoeboard=[[' ','1','2','3'],['1',' ',' ',' '],['2',' ',' ',' '],['3',' ',' ',' ']]
    print("Let's play!")
    game=False
    turn=1
    while True:
        display_board(tictactoeboard)
        if player=='x':
            print("Player X's turn...")
            player_input(tictactoeboard,player)
        else:
            print("Player O's turn...")
            player_input(tictactoeboard,player)
        if check_winX(tictactoeboard):
            display_board(tictactoeboard)
            print("Player X is the winner!")
            game=True
        if check_winO(tictactoeboard):
            display_board(tictactoeboard)
            print("Player O is the winner!")
            game=True
        if game==True:
            break
        if check_tie(tictactoeboard,turn):
            display_board(tictactoeboard)
            print("The game is a tie!")
            game=True
        if game==True:
            break
        turn+=1
        if player=='x':
            player='o'
        else:
            player='x'

play()

```
