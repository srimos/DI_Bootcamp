```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.plot(x,y)
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.scatter(x,y)
```

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.plot(x,y)
ax.set(title='Product Demand vs Price', xlabel='Price', ylabel='Demand')
fig.savefig('product_demand_vs_price.png')
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.scatter(x,y)
ax.set(title="Children's Height vs Age", xlabel='Age (years)', ylabel='Height (cm)')
fig.savefig('children_height_vs_age.png')
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs=plt.subplots(1,2)

axs[0].plot(x,y1,'tab:blue')
axs[0].set(title='y = sin(x)')

axs[1].scatter(x,y2,marker='o')
axs[1].set(title='y = cos(x)')
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs=plt.subplots(2,1)

axs[0].plot(x,y1,'tab:blue')

axs[1].scatter(x,y2,marker='o')
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs=plt.subplots(2,1)

axs[0].plot(x,y1,'tab:blue',label='y = sin(x)')
axs[0].legend(loc='upper left')

axs[1].scatter(x,y2,marker='o',label='y = cos(x)')
axs[1].legend(loc='center right')
```
